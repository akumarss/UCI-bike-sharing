```python
import pandas as pd
import numpy as np
```

```python
bikes = pd.read_csv("UCI_BikeSharing_day.csv")
bikes.head(5)
```

```python
bikes = bikes.drop("dteday", axis=1)
bikes = bikes = bikes.drop("atemp", axis=1)
bikes = bikes = bikes.drop("casual", axis=1)
bikes = bikes = bikes.drop("registered", axis=1)

bikes.columns
```

```python
feature_cols = ["temp","hum","windspeed","season","yr","mnth", "weekday", "workingday"]
category_cols = ["temp","hum","windspeed","season","yr","mnth" ]
bikes_model_df = pd.get_dummies(bikes , columns= category_cols)
bikes_model_df.columns
bikes_model_df.head(5)
```

```python

from sklearn.cross_validation import train_test_split

bike_cols = list(set(bikes_model_df.columns) - set(['cnt']))

train_X, test_X,train_Y, test_Y = train_test_split( 
                                bikes_model_df[bike_cols],
                                                 bikes_model_df.cnt,
                                                 test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(train_X, train_Y)
```

```python
y_pred = linreg.predict(test_X)
```

```python
verify = pd.DataFrame({'actual':test_Y, 'predicted':np.round(y_pred,2), 'residual':(test_Y-y_pred)})
verify.head(5)
```

```python
from sklearn import metrics
```

```python
r2score = metrics.r2_score(test_Y, y_pred)
r2score
```
